# Data Pre-Processing

We split the dataset randomly so that we have 80% as the training data and 20% as the test data.

```python
# Shuffle the data for good measure
random.shuffle(dog_food_reviews)

# Separate into texts and labels
texts = [review for review, label in dog_food_reviews]
labels = [label for review, label in dog_food_reviews]

# First, split off 10% for the test set
X_temp, X_test, y_temp, y_test = train_test_split(
    texts, labels, test_size=0.1, random_state=42
)

# Then, split the remaining 90% into 70% training and 20% validation
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=2/9, random_state=42
)
```

# Logistic Regression Model

```python
# Text -> TF-IDF features (fit only on training set)
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_val_tfidf = vectorizer.transform(X_val)
X_test_tfidf = vectorizer.transform(X_test)

# Train logistic regression
logreg_model = LogisticRegression(max_iter=1000)
logreg_model.fit(X_train_tfidf, y_train)

# Evaluate on validation set
y_val_pred = logreg_model.predict(X_val_tfidf)
print("Validation Results")
print("Accuracy:", accuracy_score(y_val, y_val_pred))
print(classification_report(y_val, y_val_pred))

# Final evaluation on test set
y_test_pred = logreg_model.predict(X_test_tfidf)
print("Test Results")
print("Accuracy:", accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

# Neural Network Model

```python
# Tokenize text
tokenizer = Tokenizer(num_words=5000, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

# Convert texts to sequences
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_val_seq = tokenizer.texts_to_sequences(X_val)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Pad sequences
X_train_pad = pad_sequences(X_train_seq, maxlen=100, padding='post')
X_val_pad = pad_sequences(X_val_seq, maxlen=100, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=100, padding='post')

# Convert labels to arrays
y_train = np.array(y_train)
y_val = np.array(y_val)
y_test = np.array(y_test)

# Build Neural Network
optimizer = Adam(learning_rate=0.0005)

nn_model = Sequential([
    Embedding(input_dim=5000, output_dim=128),
    GlobalAveragePooling1D(),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

nn_model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)

# Train Model
history = nn_model.fit(
    X_train_pad, y_train,
    validation_data=(X_val_pad, y_val),
    epochs=30,
    batch_size=32,
    callbacks=[early_stopping]
)

# Evaluate Neural Network on test set
nn_loss, nn_acc = nn_model.evaluate(X_test_pad, y_test)
print("Neural Network Results on Test Set")
print("Accuracy:", nn_acc)
```

# Results and Visualizations

Our neural network automatically stopped at 12 epochs because the accuracy stopped improving. This prevented overfitting.

```python
# Accuracy over Epochs
plt.figure(figsize=(8, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Neural Network Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# Loss over Epochs
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Neural Network Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

# Final Accuracy Comparison on Test Set

# Predict and threshold neural network outputs
y_pred_nn = nn_model.predict(X_test_pad)
y_pred_nn = (y_pred_nn > 0.5).astype(int).flatten()

# Logistic Regression Predictions already assumed to be done
y_pred_logreg = logreg_model.predict(X_test_tfidf)

# Compute test accuracies
final_logreg_acc = accuracy_score(y_test, y_pred_logreg)
final_nn_acc = accuracy_score(y_test, y_pred_nn)

# Bar chart comparison with labels on bars
plt.figure(figsize=(6, 4))
bars = plt.bar(['Logistic Regression', 'Neural Network'], [final_logreg_acc, final_nn_acc], color=['skyblue', 'salmon'])
plt.title('Final Test Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.grid(axis='y')

# Add accuracy labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.02,
        f"{height:.2f}",
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

plt.show()
```

As we can see here, the neural network slightly outperforms the logistic regression model in accuracy. However, it is very marginal and the logistic regression model is very competitive for being far lighter-weight.
